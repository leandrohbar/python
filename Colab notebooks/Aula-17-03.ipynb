```python
#INVERTENDO VARIAVEIS AULA DIA 17/03

a = int(input('Digite o valor de a = '))
b = int(input('Digite o valor de b = '))
c = a # passando 'a' para uma variavel temporaria
a = b # atribuindo 'b' a 'a'
b = c # atriuindo o valor origianal de 'a' a 'b'

#exibindo o valor da variaveis
print ('O valor das variaveis é', 'a =', a, ', b =', b) 
```

```python
#verifica se é positivo
#comandos condicionais

n = int (input ('Digite um numero inteiro = '))

if (n >= 0):
  print ('O valor é positivo')

else:
  print ('O valor é negativo')

```

```python
#calcular e mostrar o IMC da pessoa

altura = float( input ('Digite sua altura = '))
peso = float ( input ('Digite seu peso = '))
imc = peso/(altura * altura)
print (imc)

#exibir grau de IMC
if (imc <= 18.5):
  print ('abaixo do peso')
if (imc >= 18.5) and (imc <=24.9):
  print ('Peso Normal')
if (imc >= 25) and (imc <= 29.9):
  print ('Sobrespeso')
if (imc >= 30) and (imc <= 34.9):
  print ('Obesidade grau 1')
if (imc >= 35) and (imc <= 39.9):
  print ('Obesidade grau 2')
if (imc >= 40):
  print ('Obesidade grau 3 ou morbida')
```
