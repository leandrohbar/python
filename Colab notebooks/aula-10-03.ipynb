```python
pi = 3.14

#pedir para usuario digitar o raio
raio = float (input('Digite o raio da mesa '))

#calcular o raio da mesa
AreaMesa = pi * (raio * raio)

#exibir o raio da mesa
print ('A area da mesa é:', AreaMesa)
```
