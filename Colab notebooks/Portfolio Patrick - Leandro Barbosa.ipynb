```python
#exercicio 22 da lista 6 - Leandro Barbosa
#bubble sort

v = [0] * 10

for i in range(0,10):
  v[i] = (int(input('Digite um valor = ')))

count = 0
print(v)

for i in range(0, 9):
  for j in range(0, 9 - i):
    if(v[j] > v[j + 1]):
      count = v[j]
      v[j] = v[j + 1]
      v[j + 1] = count
  
print(v)
```

```python
#exercicio 28 da lista 6 - Leandro Barbosa

#pede ao usuario que digite o tamanho do vetor
colunas = int(input('Digite o numero de colunas = '))

#cria o vetor
lista = [0 for _ in range(colunas)]

#pede cada numero do vetor ao usuario
for i in range(colunas):
  lista[i] = int(input('Digite um numero = '))

#crie uma lista para fazer a rotacao
rotacao = [0 for _ in range(colunas)]

rotacao[0] = lista[-1] #traz o ultimo elemento para a 1 posicao

for i in range(colunas - 1):   #coloca cada elemento uma posicao a frente
  rotacao[i + 1] = lista[i]

lista = rotacao

print(lista)
```

```python
#exercicio 14 da lista de matriz - Leandro Barbosa

a = [[0 for _ in range(4)] for _ in range(6)]
b = [[0 for _ in range(4)] for _ in range(6)]

s = [[0 for _ in range(4)] for _ in range(6)]
d = [[0 for _ in range(4)] for _ in range(6)]

#preencher os numeros da matriz a
for i in range(6):
  for j in range(4):
    a[i][j] = int(input('Digite um numero = '))

#preencher os numeros da matriz b
for i in range(6):
  for j in range(4):
    b[i][j] = int(input('Digite um numero = '))

#criando a matriz atraves da soma
for i in range(6):
  for j in range(4):
    s[i][j] = a[i][j] + b[i][j]

#criando a matriz atraves da subtração
for i in range(6):
  for j in range(4):
    d[i][j] = a[i][j] - b[i][j]

print('a =',a,'\nb =',b,'\ns =',s,'\nd =',d)
```
