```python
#exercicio 2

num1 = int (input ('digite um numero inteiro = '))
num2 = int (input ('digite um numero inteiro = '))

if (num1 > num2):
  print ('Numero 1 maior')
elif (num1 < num2):
  print ('Numero 2 é maior')
else:
  print ('Os numeros são iguais')
```

```python
#exercicio 3

#pedir usuario para digitar um numero inteiro
num = int (input ('digite um numero inteiro = '))

#verifcar condições e printar se é ou não divisivel por 5
if (num <= 0):
  print ('Digite um numero maior que 0')

elif (num % 10 == 0) or (num % 10 == 5):
  print ('Numero multiplo de 5')

else:
  print ('Numero não divisivel por 5')
```

```python
#exercicio 4

#pedir para usuario digitar numero inteiro
num = int (input ('Digite um numero inteiro = '))

#verifcar se numero e par ou impar e printar o resultado
if (num % 2 == 0):
  print ('Numero par')
else:
  print ('Numero impar')
```
